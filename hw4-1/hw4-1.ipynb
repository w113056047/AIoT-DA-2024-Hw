```python
import pandas as pd
from pycaret.classification import *

data_train = pd.read_csv('train.csv')
data_test = pd.read_csv('test.csv')
data_train.head()
```

```python
from sklearn.model_selection import train_test_split
train_data, test_data = train_test_split(data_train.drop(['PassengerId'], axis=1), random_state=100, train_size=0.8)

train_data.head()
```

```python
# Initialize the setup
clf = setup(data=train_data,
            target='Survived',
            categorical_features = ['Pclass','Sex','Age','Fare','Cabin'])

# Compare 16 models and select the best one
best_models = compare_models(n_select=16, sort='Accuracy')

# Tune the best model
tuned_model = tune_model(best_models[0])

# Make predictions on the test data
predictions = predict_model(tuned_model, data=test_data)

# Evaluate the model
evaluate_model(tuned_model)

# Finalize the model for deployment
final_model = finalize_model(tuned_model)

# Save the model
save_model(final_model, 'titanic_model')
```
